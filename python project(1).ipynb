```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "x":["CSK", "GT", "KKR", "MI", "RR", "RCB", "SH"],
    "y":[5,1,3,5,1,1,1]
})
sns.lineplot(data=df, x = "x", y = "y", ls = "--", marker = "s")
plt.title("IPL Trophies by Team")
plt.xlabel("Teams")
plt.ylabel("Number of Trophies")
plt.show()
```

```python
import matplotlib.pyplot as plt

players = [
    "Virat Kohli", "Rohit Sharma", "Shikhar Dhawan", "David Warner", 
    "Suresh Raina", "MS Dhoni", "KL Rahul", "AB de Villiers", 
    "Chris Gayle", "Robin Uthappa", "Dinesh Karthik", "Ajinkya Rahane",
    "Faf du Plessis", "Sanju Samson"
]

runs = [8618, 7046, 6769, 6565, 5528, 5439, 5222, 5162, 4965, 4952, 4842, 4969, 4773, 4704]

plt.figure(figsize=(8, 8))

plt.pie(
    runs, 
    labels=players, 
    autopct='%1.1f%%',     
    startangle=140,        
    shadow=True
) 

plt.title('Percentage of Total Runs by Players')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

players = [
    "Virat Kohli", "Rohit Sharma", "Shikhar Dhawan", "David Warner", 
    "Suresh Raina", "MS Dhoni", "KL Rahul", "AB de Villiers", 
    "Chris Gayle", "Robin Uthappa", "Dinesh Karthik", "Ajinkya Rahane",
    "Faf du Plessis", "Sanju Samson"
]

runs = [8618, 7046, 6769, 6565, 5528, 5439, 5222, 5162, 4965, 4952, 4842, 4969, 4773, 4704]

plt.figure(figsize=(10,6))
plt.barh(players, runs, color='royalblue')
plt.title('Runs Scored by Players (Horizontal Bar Graph)')
plt.xlabel('Total Runs')
plt.ylabel('Players')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
